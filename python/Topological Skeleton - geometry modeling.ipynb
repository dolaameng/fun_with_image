<h1>Medial axis skeletonization</h1>
<p>The medial axis of an object is the set of all points having more than one closest point on the object’s boundary. It is often called the <b>topological skeleton</b>, because it is a 1-pixel wide skeleton of the object, with the same connectivity as the original object.
</p>
<p>Here, we use the medial axis transform to compute the width of the foreground objects. As the function <code>medial_axis (skimage.morphology.medial_axis)</code> returns the distance transform in addition to the medial axis (with the keyword argument <code>return_distance=True</code>), it is possible to compute the distance to the background for all points of the medial axis with this function. This gives an estimate of the local width of the objects.
</p>

```python
import numpy as np
from scipy import ndimage
from skimage.morphology import medial_axis
```

```python
## generate synthetic binary data - binary microstructure with blobs
## param (optional): linear size of the returned image
def microstructure(sz = 256):
    n = 5
    x, y = np.ogrid[0:sz, 0:sz]
    mask_outer = (x - sz/2) ** 2 + (y - sz/2) ** 2 < (sz / 2)**2
    mask = np.zeros((sz, sz))
    generator = np.random.RandomState(1)
    points = sz * generator.rand(2, n**2)
    mask[(points[0]).astype(np.int), (points[1]).astype(np.int)] = 1
    mask = ndimage.gaussian_filter(mask, sigma = sz / (4. * n))
    return mask > mask.mean()

data = microstructure(sz = 64)
figure(figsize= (6, 6))
imshow(data, cmap=cm.gray)
```

```python
## compute the medial axis (skeleton) and the distance transform
skel, distance = medial_axis(data, return_distance=True)
## distance to the background for pixels of the skeleton
## i.e., the distance from the medial axis points to the boundaries of objects
dist_on_skel = distance * skel
```

```python
figure(figsize=(6, 6))
imshow(dist_on_skel, cmap=cm.spectral)
contour(data, [0.5], colors = 'w') # show the boundary for binary image
```
