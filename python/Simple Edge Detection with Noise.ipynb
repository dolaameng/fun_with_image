<h1>Edge Detection using Canny Edge Detector</h1>
<h2>Problem: Detect the edge of objects in a NOISY image. Thin down the blurred edges to 1-pixel curves</h2>
<p>Tech: The Canny filter is a multi-stage edge detector. It uses a filter based on the <em>derivative of a Gaussian</em> in order to compute the intensity of the gradients.The Gaussian reduces the effect of noise present in the image. Then, potential edges are thinned down to 1-pixel curves by <em>removing non-maximum pixels of the gradient magnitude </em>. Finally, edge pixels are kept or removed using <em>hysteresis thresholding</em> on the gradient magnitude.</p>
<p>The Canny has three adjustable parameters: the width of the Gaussian (the noisier the image, the greater the width), and the low and high threshold for the hysteresis thresholding.</p>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from skimage import filter
```

```python
## generate the noisy image of an square object
im = np.zeros((128, 128))
im[32:-32, 32:-32] = 1
im = ndimage.rotate(im, 15, mode='constant') # rotation
im = ndimage.gaussian_filter(im, 4) # blurred edge
im += 0.2 + np.random.random(im.shape) # noise
plt.imshow(im, cmap=cm.jet)
```

```python
## first try  Canny filter with different width parameters
for sigma in xrange(1, 10):
    figure()
    edges = filter.canny(im, sigma=sigma)
    plt.imshow(edges, cmap=cm.gray)
    plt.title('$\sigma=%d$' % (sigma,))
```
