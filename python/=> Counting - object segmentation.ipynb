<h1>Counting Nuclei in DNA</h1>
<p>
    Our first task will be to take this image and count the number of nuclei.
</p>

```python
import numpy as np
import scipy as sp
import pylab, pymorph, mahotas
from scipy import ndimage
```

```python
## DNA image
## pylab show the image by default with colormap = heatmap
## and it contrast-stretches the images before displaying them
dna = mahotas.imread('dna.jpeg')
pylab.imshow(dna)
print dna.shape, dna.dtype, dna.max(), dna.min()
## show it again as a grayscale - because there is no color channels
figure()
pylab.imshow(dna, cmap=cm.gray)
```

<p>How many nuclei are there?</p>
<p>It turns out counting in such a case is not THAT easy - I have tried it 5 times, 3 times got 19, 1 time got 20, 1 time got 18</p>
<p>Actually, there are 21, two on the edge, which are very hard to recognize</p>

<h2>First Idea - Counting by Thresholding</h2>
<p>Thresholding by <a href="http://en.wikipedia.org/wiki/Otsu's_method">OTSU method</a> -- automatic histogram shape-based image thresholding</p>
<p><em>It assumes that the graylevel image contains two classes of pixels</em></p>

```python
## find the threshold
T = mahotas.thresholding.otsu(dna)
## binary image as a mask
pylab.imshow(dna > T, cmap=cm.gray)
```

<p>The problem here is that the objects recognized are not smooth enough, that could be difficult when counting them</p>

<h2>Somooth the image using a Gaussian filter</h2>
<p>There are several ways of solving above "too many small objects" problem. One of the simplest one is to smooth the image a big using a Gaussian filter - that essentially reduces the noise level</p>

```python
## filtered image
dnaf = ndimage.gaussian_filter(dna, 8)
## do the otsu filtering again
T = mahotas.thresholding.otsu(dnaf)
pylab.imshow(dnaf > T, cmap=cm.gray)
```

<p>Now much better result - the nuclei objects have been MERGED, some of them are still touching, but the result looks better in overall.</p>

<h2>Counting objects by labelling</h2>

```python
## labeled object is an image (ndarray) of the exact same size as the original image
## the valute in the labeled image is the label of the object at that position, from 0 (background)
## to nr_objects
labeled, nr_objects = ndimage.label(dnaf > T)
print 'Number of objects found: ', nr_objects-1
pylab.imshow(labeled)
```

<p>The result is much clearer now, but there are still nuclei glued together. Let's try better!</p>

<h2>Counting by Segmenting the Image</h2>
<p>A simple and traditional trick: </p>
<ol>
    <li>smooth the image by filters</li>
    <li>find regional maxima </li>
    <li>use the regional maxima as SEEDS for <a href="http://en.wikipedia.org/wiki/Watershed_(image_processing)">WATERSHED</a></li>
</ol>
<p>
    A grey-level image may be seen as a topographic relief, where the grey level of a pixel is interpreted as its altitude in the relief. A drop of water falling on a topographic relief flows along a path to finally reach a local minimum. Intuitively, the watershed of a relief correspond to the limits of the adjacent catchment basins of the drops of water.
</p>

<h3>Finding the seeds as regional maxima</h3>

```python
## filtering with Gaussian filter
dnaf = ndimage.gaussian_filter(dna, 8)
## find regional max
rmax = pymorph.regmax(dnaf)
pylab.imshow(rmax, cmap=cm.gray)
## apply binary overlay
figure()
pylab.imshow(pymorph.overlay(dnaf, rmax)) # by default as RED
```

<p>The result does NOT look very good - multiple maxima in single region</p>
<p>One explaination would be that the noise level is still too high - which means we have to filter the image more</p>
<p>And this is an EXPLORATIVE STUDY - we have to fiddle around with bigger sigma values for Gaussian filters</p>

```python
dnaf = ndimage.gaussian_filter(dna, 20) # 16.5
rmax = pymorph.regmax(dnaf)
figure(figsize=(10, 10))
pylab.imshow(pymorph.overlay(dnaf, rmax))
## we can count the number of seeds
seeds, nr_nuclei = ndimage.label(rmax)
print nr_nuclei
```

<p>Not bad, maybe good enough for watersheding</p>

<h3>Watershed</h3>
<p>We are going to apply watershed to the distance transform of the thresholded image</p>

```python
## binarize the image based on the otsu threshold
T = mahotas.thresholding.otsu(dnaf)
## calcualte the distance transform of pixels
dist = ndimage.distance_transform_edt(dnaf > T)
## normalize the distance -- contrast stretched
dist = dist.max() - dist
dist -= dist.min()
dist = dist / float(dist.ptp()) * 255
dist = dist.astype(np.uint8)
pylab.imshow(dist)
```

```python
## after contrast stretching the dist image
## use the watershed
nuclei = pymorph.cwatershed(dist, seeds)
pylab.imshow(nuclei)
```
