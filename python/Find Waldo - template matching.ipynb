```python
from pylab import imshow
import mahotas
```

```python
## read the image, convert it to a 3D matrix
store = mahotas.imread('./DepartmentStore.jpg')
## convert integer (r,g,b) to float for better precision of arithmetics
figure(figsize=(12,12))
store = store.astype('float')
imshow(store.astype('uint8'))
print store.shape
```

```python
## extract red, green, and blue channel
r, g, b = store.transpose((2, 0, 1))
print r.shape, g.shape, b.shape
## get the white channel as the mean of r, g, b
w = store.mean(2)
print w.shape
```

```python
## build the pattern that we are looking for
## Walley's shirt - a pattern of -1, -1, +1, +1 on the vertical axis
pattern = np.ones((24, 16), float)
for i in xrange(2):
    pattern[i::4] = -1
imshow(pattern, cm.binary)
```

```python
## build Wally's shirt by convolving red and white with above pattern
wally = mahotas.convolve(r - w, pattern)
imshow(wally)
```

```python
## when convolved, the shirt will give a strong response as max value
## dilate it to make it visible
mask = (wally == wally.max())
mask = mahotas.dilate(mask, np.ones((48, 24)))
imshow(mask)
```

```python
## house-keeping - with the mask, tone down the whole image, except the region of interest
store -= .8 * store * ~mask[:, :, None]
figure(figsize=(12, 12))
imshow(store.astype('uint8'))
```
