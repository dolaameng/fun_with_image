```python
from PIL import Image
import numpy as np
import fractions
from pylab import imshow
import mahotas
```

```python
## load the image and convert it to a gray-scale
shredded = np.asarray(Image.open('shredded.png').convert('L'))
print shredded.shape
print shredded.dtype
imshow(shredded, cmap=cm.gray)
```

```python
## compute the mean value of each column of pixels and the difference
## between adjancent columns
column_means = np.mean(shredded, axis = 0)
column_diffs = np.diff(column_means)
print column_diffs.shape
```

```python
## IDEA - the greatest differences of mean column ixel should occur at boundaries
##      - and the indices of these boundaries are multiples of the shred width.
## Iterative search - we'll raise the threshold of waht we consider to be a big difference 
## until all the boundaries occur at multiples of some reasonably large width, i.e., 
## until they share a greatest common denominator > some minimal shred width.
threshold, width = 1, 0
while width < 5 and threshold < 255:
    boundaries = [i for i, diff in enumerate(column_diffs, 1) if diff > threshold]
    width = reduce(lambda x, y: fractions.gcd(x, y), boundaries) if boundaries else 0
    threshold += 1
print 'shred width', width
print 'boundaries index', boundaries
```

```python
## find shreds and sub-images of different shreds
shreds = range(shredded.shape[1] / width)
bounds = [(shredded[:, width*shred], shredded[:, width*(shred+1)-1]) for shred in shreds]
print len(bounds)
```

```python
## build the similiarity matrix of different shreds
D = [[np.linalg.norm(bounds[s2][1] - bounds[s1][0]) 
            if s1 != s2 else numpy.Infinity 
                            for s2 in shreds] 
                                for s1 in shreds]
```

```python
## find neighbors based on the similiarity matrix
neighbors = [np.argmin(D[shred]) for shred in shreds]
print list(enumerate(neighbors))
```

```python
## find the original order of strips by chain the neighbors in a loop
## function to chain neighors of shreds - return the max length for a loop and the chained loop
def chain(start, neighbors):
    loop = [start]
    while True:
        current_node = loop[-1]
        next_node = neighbors[current_node]
        if next_node in loop: break
        loop.append(next_node)
    return len(loop), loop

print [chain(start, neighbors)[0] for start in shreds]
new_order = max([chain(start, neighbors) for start in shreds])[1]
## reverse it
new_order = new_order[::-1]
```

```python
## find the most likely order (with max length) of the shreds
print new_order, len(new_order)
```

```python
## reconstruct the image
shredded = Image.open('shredded.png')
unshredded = Image.new('RGBA', shredded.size)
for to, fr in enumerate(new_order):
    source = shredded.crop((fr*width, 0, (fr+1)*width, shredded.size[1]))
    destination = (to * width, 0)
    unshredded.paste(source, destination)
figure(figsize=(10, 10))
imshow(np.array(unshredded))
```
