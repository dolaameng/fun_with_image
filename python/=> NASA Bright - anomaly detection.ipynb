```python
from pylab import imshow
import mahotas
from pandas import DataFrame, Series
import pandas as pd
```

```python
## load image matrix
nasa = mahotas.imread('./nasa.jpg')
nasa = nasa.astype('float') / 255.
figure(figsize = (10, 10))
imshow(nasa)
print nasa.shape
```

```python
## look at the sand area only - a HOPEFULLY clean area
## OUR REFERENCE
sand = nasa[:300, 1000:, :]
figure(figsize=(6, 6))
imshow(sand)
print sand.shape
```

```python
## find the typical sand color values - mean and sd
## ANALYSIS OF REFERENCE in three different color channels (RGB)
sand_mean = np.array([np.mean(sand[:,:,i]) for i in range(3)])
sand_std = np.array([np.std(sand[:, :, i]) for i in range(3)])
print sand_mean
print sand_std
```

```python
## how much do all colors in the whole picture deviate from sand colors
colorz = np.array([(nasa[:,:,i] - sand_mean[i]) / sand_std[i] for i in range(3)])
colorz = np.transpose(colorz, axes = (1, 2, 0))
print colorz.shape
imshow(colorz)
```

```python
density = np.sum(colorz, axis = 2) ## sum along all colors (RGB)
print density.shape
density_plot = Series(density.flatten())
density_plot.plot(kind = 'kde')
```

<p>There are two obvious peaks in the spectrum, one should be sand (background), one should be bright objects (machine components)</p>
